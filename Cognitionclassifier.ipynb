# **Sampling**

```python
data
```

```python
data_minority = data[data['Cognition'] == 'DM']
```

```python
data_minority.shape
```

```python
data_majority = data[data['Cognition'].isin(['CN', 'MCI'])]
```

```python
data_majority.shape
```

```python
# Downsample majority class
from sklearn.utils import resample
data_majority_downsampled = resample(data_majority,
                                 replace=False,
                                 n_samples=400)
#Upsample minority class
data_minority_upsampled = resample(data_minority,
                                 replace=True,
                                 n_samples=300)
# Combine minority class with downsampled majority class
data_sampled = pd.concat([data_majority_downsampled, data_minority_upsampled])
```

```python
data_sampled.shape
```

```python
class_sampled = data_sampled.Cognition.value_counts()
class_sampled
```

# *Logistic Regression for classification*

```python
#importing libraries needed
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model
```

```python
# assigning different features to x variable
x = data_sampled.drop(columns='Cognition')
# assigning target column to y variable
y = data_sampled.iloc[:,17]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
#Standardize the features
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
#log regression
lm = linear_model.LogisticRegression(multi_class='ovr', solver='liblinear',class_weight='balanced')
lm.fit(x_train_scaled, y_train)
```

```python
#checking test data
y_pred = lm.predict(x_test_scaled)
```

```python
y_pred
```

```python
print("Accuracy of logistic Regression: ",lm.score(x_test_scaled,y_test))
```

```python
from sklearn.metrics import confusion_matrix
cf_mat = confusion_matrix(y_test, y_pred)
print('Confusion matrix')
print(cf_mat)
```

# **Random Forest**

```python
# importing random forest classifier from assemble module
from sklearn.ensemble import RandomForestClassifier
# creating a RF classifier
clf = RandomForestClassifier(n_estimators = 100)
```

```python
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
clf.fit(x_train_scaled, y_train)
```

```python
# performing predictions on the test dataset
y_pred = clf.predict(x_test_scaled)
```

```python
y_pred
```

```python
# metrics are used to find accuracy or error
from sklearn import metrics
print()

# using metrics module for accuracy calculation
print("ACCURACY OF THE Random forest: ", metrics.accuracy_score(y_test, y_pred))
```

```python
# passing actual and predicted values
cm = confusion_matrix(y_test, y_pred)
# true Write data values in each cell of the matrix
sns.heatmap(cm,annot=True)
```
